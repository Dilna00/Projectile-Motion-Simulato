```python
# Projectile Motion Simulator

This interactive Jupyter Notebook simulates and visualizes the trajectory of a projectile under gravity, with options for air resistance. You can adjust parameters and see real-time results.
```

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import FloatSlider, Checkbox, VBox, HBox, Output
from IPython.display import display, Markdown

# Check if widgets and plotting work
print("Environment ready for interactive Python coding.")
```

```python
# Interactive widgets for projectile parameters
initial_speed_slider = FloatSlider(value=20, min=1, max=100, step=0.5, description='Speed (m/s)')
angle_slider = FloatSlider(value=45, min=0, max=90, step=1, description='Angle (°)')
mass_slider = FloatSlider(value=1, min=0.1, max=10, step=0.1, description='Mass (kg)')
g_slider = FloatSlider(value=9.81, min=1, max=20, step=0.01, description='Gravity (m/s²)')
drag_checkbox = Checkbox(value=False, description='Air Resistance')
drag_coeff_slider = FloatSlider(value=0.1, min=0.01, max=1, step=0.01, description='Drag Coeff.')

widgets_box = VBox([
    initial_speed_slider,
    angle_slider,
    mass_slider,
    g_slider,
    HBox([drag_checkbox, drag_coeff_slider])
])
display(widgets_box)
```

```python
# Function to compute trajectory (with/without air resistance)
def compute_trajectory(v0, angle_deg, g, mass, drag, drag_coeff):
    angle = np.radians(angle_deg)
    dt = 0.01
    t = 0
    x, y = 0, 0
    vx = v0 * np.cos(angle)
    vy = v0 * np.sin(angle)
    X, Y, T = [x], [y], [t]
    while y >= 0:
        if drag:
            v = np.sqrt(vx**2 + vy**2)
            Fd = drag_coeff * v**2
            ax = -Fd * vx / (mass * v)
            ay = -g - (Fd * vy / (mass * v))
        else:
            ax = 0
            ay = -g
        vx += ax * dt
        vy += ay * dt
        x += vx * dt
        y += vy * dt
        t += dt
        X.append(x)
        Y.append(y)
        T.append(t)
        if t > 100:  # prevent infinite loop
            break
    return np.array(X), np.array(Y), np.array(T)

# Output widget for plot
plot_out = Output()

def update_plot(*args):
    with plot_out:
        plot_out.clear_output()
        v0 = initial_speed_slider.value
        angle = angle_slider.value
        mass = mass_slider.value
        g = g_slider.value
        drag = drag_checkbox.value
        drag_coeff = drag_coeff_slider.value if drag else 0.0
        X, Y, T = compute_trajectory(v0, angle, g, mass, drag, drag_coeff)
        plt.figure(figsize=(8,4))
        plt.plot(X, Y)
        plt.title('Projectile Trajectory')
        plt.xlabel('Distance (m)')
        plt.ylabel('Height (m)')
        plt.grid(True)
        plt.ylim(bottom=0)
        plt.show()
        # Show stats
        max_height = np.max(Y)
        range_ = np.max(X)
        time_of_flight = T[len(Y)-1]
        display(Markdown(f"**Max Height:** {max_height:.2f} m  "))
        display(Markdown(f"**Range:** {range_:.2f} m  "))
        display(Markdown(f"**Time of Flight:** {time_of_flight:.2f} s"))

# Link widgets to update function
for w in [initial_speed_slider, angle_slider, mass_slider, g_slider, drag_checkbox, drag_coeff_slider]:
    w.observe(update_plot, names='value')

update_plot()
display(plot_out)
```
